```python
## abs()
print(abs(-10))
f = abs
print(f(-1))
# 内置函数是可以被赋值给其他变量的，同样也可以将其他对象赋值给内置函数
abs = id
print(abs(1))
# 不要使用和内置函数重名的变量名，这会让代码混乱，容易发生难以排查的错误

## all()
print(all([1, 2, 3]))
print(all([1, 0, 0]))

## any()
print(any([0, 1, 0]))
print(any([0, 0, 0]))

## ascii()
s = 'haha'
print(ascii(s))

## bytes() 将对象转换成字节类型
i = 2
print(bytes(i))
s = '张三'
m = bytes(s, encoding='utf-8')
print(m)
s = 'haha'
print(bytes(s, encoding='utf-8'))

## dir()  显示对象所有的属性和方法
dir()

## divmod()  同时返回商和余数的元组
print(divmod(10, 3))

## enumerate()  在迭代对象的时候，额外提供一个序列号的输出.enumerate(li,1)中的1表示从1开始序号，默认从0开始。注意，第二个参数才是你想要的序号开始，不是第一个参数。
dic = {
    'k1': 'v1',
    'k2': 'v2',
    'k3': 'v3',
}
for i, k in enumerate(dic, 1):
    print(str(i) + '-->' + k)

## eval() 将字符串直接解读并执行
print(eval('6*8'))

## globals() 列出当前环境下所有的全局变量
globals()

## hash()  为不可变对象，例如字符串生成哈希值的函数
print(hash((1, 2, 3)))

## help() 返回对象的帮助文档
a = [1, 2, 3]
# help(a)

## id() 返回对象的内存地址,常用来查看变量引用的变化，对象是否相同等
id(0)
print(id(True))

## isinstance()  判断一个对象是否是某个类的实例
isinstance('gg', str)

## iter()    制造一个迭代器，使其具备next()能力
lis = [1, 2, 3]
l = iter(lis)
n = next(l)
n

## locals() 返回当前可用的局部变量
locals()

## next()  通过调用迭代器的__next__()方法，获取下一个元素

## repr()  用对象所属类的__repr__方法，与print功能类似
s = 'hahaha'
print(repr(s))

## reversed()
print(type(reversed))
a = reversed([1, 2, 3, 4, 5])
print('函数对象:', a)
print('反向结果:', list(a))

## round() 四舍五入
print(round(1.5))

## map()
li = [1,2,3]
data = map(lambda x :x*100,li)  # 这里直接使用了一个匿名函数

print(type(data))       # 返回值是一个map对象，它是个迭代器。
data = list(data)       # 可以用list方法将map对象中的元素全部生成出来，保存到一个列表里。
print(data)

## filter()
def f1(x):
    if x > 3:
        return True
    else:
        return False
li = [1,2,3,4,5]
data = filter(f1,li)
print(type(data))
print(list(data))

## zip()   组合对象，将对象逐一配对
lis1 = [1, 2, 3, 4]
lis2 = ['a', 'b', 'c', 'd']
s = zip(lis1, lis2)
print('函数对象:', s)
print('结果:', list(s))
### 如果对象的长度不一致呢？多余的会被抛弃！以最短的为基础！
lis1 = [1, 2, 3]
lis2 = ['a', 'b', 'c', 'd']
s = zip(lis1, lis2)
print('函数对象:', s)
print('结果:', list(s))

## sorted() 排序方法。有key和reverse两个重要参数。
sorted([36, 5, -12, 9, -21])
### 指定排序的关键字。关键字必须是一个可调用的对象
sorted([36, 5, -12, 9, -21], key=abs)
### 指定按反序排列
sorted(['bob', 'about', 'Zoo', 'Credit'], key=str.lower, reverse=True)
```
